```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("GSDP.csv")
print(df.head())
kerala = df[df['State/Uts'] == 'Kerala'].iloc[0]
years = df.columns[1:11]
gsdp_values = kerala[1:11].values.astype(float)
plt.figure(figsize=(10,5))
plt.plot(years, gsdp_values, marker='o', color='green', linewidth=2)
plt.title("Kerala: Gross State Domestic Product (GSDP) at Current Prices")
plt.xlabel("Financial Year")
plt.ylabel("GSDP (₹ Crores)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("GAT.xls")
df_subset = df[['Emissions from Fossil Fuel Burning (Million Tonnes of Carbon)', 'Carbon Dioxide (Parts Per Million)']]
print(df_subset.head())
# Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(df_subset['Emissions from Fossil Fuel Burning (Million Tonnes of Carbon)'], df_subset['Carbon Dioxide (Parts Per Million)'], color='green')
plt.title('Fossil Fuel Emmisions vs CO2 Levels')
plt.xlabel('Fossil Fuel Emmisions')
plt.ylabel('CO2(ppm)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("sectors.csv", index_col=0)

df_subset = df.iloc[:3, :40]
print(df_subset.head())

plt.figure(figsize=(8,6))
df_subset.T.boxplot()
plt.title("Distribution of Quarterly Values by Sector in 2011-2021")
plt.ylabel("Value")
plt.xlabel("Sector")
plt.show()


plt.figure(figsize=(10,6))
for sector in df_subset.index:
    plt.plot(df_subset.columns, df_subset.loc[sector], marker='o', label=sector)

plt.title("Quarterly Trends Across Sectors")
plt.xlabel("Quarter")
plt.ylabel("Value")
ticks_to_show = range(0, len(df_subset.columns), 4)
df_subset.columns = [col.split('-', 1)[1] for col in df_subset.columns]
plt.xticks(ticks=ticks_to_show, labels=[df_subset.columns[i] for i in ticks_to_show])
# plt.xticks([])
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
```
